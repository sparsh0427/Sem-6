QUESTION 1: MATRIX MULTIPLICATION

```
dim = 100
matrix_a = np.random.randint(1,1000,(dim,dim))
matrix_b = np.random.randint(1,1000,(dim,dim))
```

SERIAL OPERATION

```
%%time 
matrix_c = np.zeros((dim,dim))
matrix_c = matrix_c.astype(int)

for i in range(dim): 
    for j in range(dim): 
        for k in range(dim): 
            matrix_c[i][j] += matrix_a[i][k] * matrix_b[k][j] 

print(matrix_c)
print("Time for Serial Operation: ")
```

MULTITHREADING

```
%%time
n=matrix_a.shape[0]

def work_1(matrix_a, matrix_b, matrix_c, i):
    for j in range(n):
        for k in range(n):
            matrix_c[i][j] += matrix_a[i][k] * matrix_b[k][j]
            

threads = [None]*n
matrix_c = np.zeros((n,n))

for i in range(n):
  threads[i] = threading.Thread(target = work_1, args = (matrix_a, matrix_b, matrix_c, i))
  threads[i].start()

for i in range(n):
  threads[i].join()

print(matrix_c)
print("Time for Multithreading: ")
```

QUESTION 3: Array Sum

```
lst = np.random.randint(500,size = 5000000)
```

USING LOOP

```
%%time
sum = 0
for i in range(len(lst)):
    sum += lst[i]
print(sum)
print("Time for Serial Operation: ")
```

MULTITHREADING

```
%%time

def arr_sum(lst):
    sum = 0
    for i in range(len(lst)):
        sum += lst[i]
    return sum

que = queue.Queue()
threads_list = list()

n=len(lst)
total_sum = 0

t = threading.Thread(target=lambda q, arg1: q.put(arr_sum(arg1)), args=(que, lst[:(n//2)]))
t.start()
threads_list.append(t)

t2 = threading.Thread(target=lambda q, arg1: q.put(arr_sum(arg1)), args=(que, lst[(n//2):]))
t2.start()
threads_list.append(t2)

for t in threads_list:
    t.join()

while not que.empty():
    total_sum += que.get()
print(total_sum)
print("Time for Multithreading: ")
```

QUESTION 4: Prime Numbers

```
start = 1
end = 50000
```

SERIAL OPERATION

```
%%time
lst=[]
for num in range(start, end + 1):
       for i in range(2, num):
           if (num % i) == 0:
               break
       else:
           lst.append(num)
print(lst)
print("Time for Serial Operation: ")
```

MULTITHREADING

```
%%time

def prime(low,high,out):
    res=[]
    for num in range(low, high + 1):
       for i in range(2, num):
           if (num % i) == 0:
               break
       else:
           res.append(num)
    out.put(res)

que = queue.Queue()
threads_list = list()

total = []

t1 = threading.Thread(prime(start, end//2, que))
t1.start()
threads_list.append(t1)

t2 = threading.Thread(prime(start//2, end, que))
t2.start()
threads_list.append(t2)

for t in threads_list:
    t.join()

for i in que.queue:
    total +=i
print(total)
print("Time for Multithreading: ")
```
